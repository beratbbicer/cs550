```python
import numpy as np
from numpy import loadtxt
from scipy.special import expit
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 7]

train1 = loadtxt("train1", delimiter="\t")
test1 = loadtxt("test1", delimiter="\t")
train2 = loadtxt("train2", delimiter="\t")
test2 = loadtxt("test2", delimiter="\t")

train1_x, train1_y = train1[:, 0], train1[:, 1]
train2_x, train2_y = train2[:, 0], train2[:, 1]
test1_x, test1_y = test1[:, 0], test1[:, 1]
test2_x, test2_y = test2[:, 0], test2[:, 1]

# train1_x = train1_x / np.linalg.norm(train1_x)
# train2_x = train2_x / np.linalg.norm(train2_x)
# test1_x = test1_x / np.linalg.norm(test1_x)
# test2_x = test2_x / np.linalg.norm(test2_x)

hidden_units = 16
t1 = np.random.uniform(0, 1, hidden_units)
t2 = np.random.uniform(0, 1, hidden_units)
t3 = np.random.uniform(0, 1, hidden_units)
t4 = np.random.uniform(0, 1, 2)
```

```python
def mlp_regression(x, y, lr, w_ih_org, w_ho_org, bias_org, max_epochs, epsilon, hidden_units, momentum, stochastic = True):
    if max_epochs <= 0 and epsilon < 0:
        return
    
    # x = np.copy(x_org)
    # y = np.copy(y_org)
    w_ih = np.copy(w_ih_org)
    w_ho = np.copy(w_ho_org)
    bias = np.copy(bias_org)
    loss_hist = []
    epochs = 0

    while True:
        delta_w_ih = np.zeros(hidden_units)
        delta_bias = np.zeros(hidden_units)
        delta_w_ho = np.zeros(hidden_units)
        prev_delta_w_ih = np.zeros(hidden_units)
        prev_delta_bias = np.zeros(hidden_units)
        prev_delta_w_ho = np.zeros(hidden_units)
        epoch_loss = 0
        
        for j in range(len(x)):
            h = expit(np.multiply(x[j], w_ih) + bias) # sigmoid
            y_hat = np.dot(h, w_ho)
            error = y_hat - y[j]
            epoch_loss += 0.5 * (error**2)
            
            d_w_ho = error * h
            d_bias = np.multiply(error * w_ho, np.multiply(h, 1-h))
            d_w_ih = d_bias * x[j]
            
            if stochastic == True:
                delta_w_ih = lr * d_w_ih * (1-momentum) + momentum * delta_w_ih
                delta_bias = lr * d_bias * (1-momentum) + momentum * delta_bias
                delta_w_ho = lr * d_w_ho * (1-momentum) + momentum * delta_w_ho
                w_ih -= delta_w_ih
                bias -= delta_bias
                w_ho -= delta_w_ho
            else:
                delta_w_ih += d_w_ih
                delta_bias += d_bias
                delta_w_ho += w_ho
        
        loss_hist.append(epoch_loss / len(x))
        
        if stochastic == False:
            delta_w_ih = lr * delta_w_ih * (1-momentum) + momentum * prev_delta_w_ih
            delta_bias = lr * delta_bias * (1-momentum) + momentum * prev_delta_bias
            delta_w_ho = lr * delta_w_ho * (1-momentum) + momentum * prev_delta_w_ho
            w_ih -= delta_w_ih
            bias -= delta_bias
            w_ho -= delta_w_ho
            prev_delta_w_ih = delta_w_ih
            prev_delta_bias = delta_bias
            prev_delta_w_ho = delta_w_ho
        
        epochs += 1     
        if epochs > max_epochs or np.mean(np.array(loss_hist)) < epsilon:
            break

    plt.plot(np.arange(1, epochs+1), loss_hist)
    plt.title("Average loss over epochs")
    plt.show(block=True)
    return w_ih, w_ho, bias, np.array(loss_hist)

def get_predictions(features, w_ih, w_ho, bias):
    y = []
    for x in features:
        h = expit(np.multiply(x, w_ih) + bias)
        y_hat = np.dot(h, w_ho)
        y.append(y_hat)
    return np.array(y)
```

```python
def linear_regression(x, y, w_org, lr, max_epochs, epsilon, momentum, stochastic = True):
    if max_epochs <= 0 and epsilon < 0:
        return
    # x = np.copy(x_org)
    # y = np.copy(y_org)
    w = np.copy(w_org)
    prev_delta_w = np.array([0, 0])
    epochs = 0
    loss_hist = []
    
    while True:
        delta_w = np.zeros(2)
        epoch_loss = 0
        
        for j in range(len(x)):
            y_hat = x[j] * w[0] + w[1]
            error = y_hat - y[j]
            epoch_loss += 0.5 * (error**2)
            d_w = np.array([error * x[j], error])
            
            if stochastic == True:
                delta_w = d_w * lr * (1 - momentum) + delta_w * momentum
                w -= delta_w
            else:
                delta_w += d_w
        loss_hist.append(epoch_loss / len(x))
        
        if stochastic == False:
            delta_w = delta_w * lr * (1 - momentum) + prev_delta_w * momentum
            w -= delta_w
            prev_delta_w = delta_w

        epochs += 1   
        if epochs > max_epochs or np.mean(np.array(loss_hist)) < epsilon:
            break
            
    plt.plot(np.arange(1, epochs+1), loss_hist)
    plt.title("Average loss over epochs")
    plt.show(block=True)
    return w, np.array(loss_hist)
```

```python
# DO NOT USE BATCH, USE STOCHASTIC
print("Train 1")
print("---- MLP")
w_ih, w_ho, bias, loss_hist_mlp = mlp_regression(train1_x, train1_y, 0.001, t1, t2, t3, max_epochs = 1000, epsilon = 200, hidden_units = hidden_units, momentum = 0.9)
print("---- RL")
w, loss_hist_lr = linear_regression(train1_x, train1_y, t4, 0.001, max_epochs = 100, epsilon = 700, momentum = 0.9)

print("||| Statistics |||")
y_pred_mlp_1 = get_predictions(test1_x, w_ih, w_ho, bias)
y_pred_lr_1 = np.array([x * w[0] + w[1] for x in test1_x])
print("---- MLP")
print("avg. loss: " + str(round(np.mean(loss_hist_mlp), 2)) + "\tstddev: " + str(round(np.std(loss_hist_mlp), 2)))
print("Test loss: " + str(round(np.mean(np.power(y_pred_mlp_1 - test1_y, 2)), 2)))
print("---- RL")
print("avg. loss: " + str(round(np.mean(loss_hist_lr), 2)) + "\tstddev: " + str(round(np.std(loss_hist_lr), 2)))
print("Test loss: " + str(round(np.mean(np.power(y_pred_lr_1 - test1_y, 2)), 2)))

print("\n\nTrain 2")
print("---- MLP")
w_ih_2, w_ho_2, bias_2, loss_hist_mlp_2 = mlp_regression(train2_x, train2_y, 0.001, t1, t2, t3, max_epochs = 100, epsilon = 82000, hidden_units = hidden_units, momentum = 0.9)
print("---- RL")
w_2, loss_hist_lr_2 = linear_regression(train2_x, train2_y, t4, 0.0001, max_epochs = 35, epsilon = 100100, momentum = 0.9)

print("||| Statistics |||")
y_pred_mlp_2 = get_predictions(test2_x, w_ih_2, w_ho_2, bias_2)
y_pred_lr_2 = np.array([x * w_2[0] + w_2[1] for x in test2_x])
print("---- MLP")
print("avg. loss: " + str(round(np.mean(loss_hist_mlp_2), 2)) + "\tstddev: " + str(round(np.std(y_pred_mlp_2), 2)))
print("Test loss: " + str(round(np.mean(np.power(y_pred_mlp_2 - test2_y, 2)), 2)))
print("---- RL")
print("avg. loss: " + str(round(np.mean(loss_hist_lr_2), 2)) + "\tstddev: " + str(round(np.std(loss_hist_lr_2), 2)))
print("Test loss: " + str(round(np.mean(np.power(y_pred_lr_2 - test2_y, 2)), 2)))
```

```python
def plot_mlp_hu(x_org, w_ih_org, w_ho_org, bias_org, hidden_units):
    x = np.sort(np.copy(x_org))
    w_ih = np.copy(w_ih_org)
    w_ho = np.copy(w_ho_org)
    bias = np.copy(bias_org)
    h_values = []
    
    for j in range(len(x)):
        h_values.append(expit(np.multiply(x[j], w_ih) + bias))
    
    h_values = np.asarray(h_values)
    for i in range(hidden_units):
        plt.plot(x, h_values[:, i])
    plt.title("MLP Hidden Units")

plt.subplot(2, 3, 1)
p = train1_x.argsort()
plt.scatter(train1_x[p], train1_y[p])
plt.plot(train1_x[p], train1_x[p] * w[0] + w[1], c="r")
plt.title("Linear Regression")

plt.subplot(2, 3, 2)
plt.scatter(train1_x[p], train1_y[p])
plt.plot(train1_x[p], get_predictions(train1_x, w_ih, w_ho, bias)[p], c="r")
plt.title(str(hidden_units) + "HU MLP")

plt.subplot(2, 3, 3)
plot_mlp_hu(train1_x, w_ih, w_ho, bias, hidden_units)

plt.subplot(2, 3, 4)
p = train2_x.argsort()
plt.scatter(train2_x[p], train2_y[p])
plt.plot(train2_x[p], train2_x[p] * w_2[0] + w_2[1], c="r")
plt.title("Linear Regression")

plt.subplot(2, 3, 5)
plt.scatter(train2_x[p], train2_y[p])
plt.plot(train2_x[p], get_predictions(train2_x, w_ih_2, w_ho_2, bias_2)[p], c="r")
plt.title(str(hidden_units) + "HU MLP")

plt.subplot(2, 3, 6)
plot_mlp_hu(train2_x, w_ih_2, w_ho_2, bias_2, hidden_units)
plt.show()
```
